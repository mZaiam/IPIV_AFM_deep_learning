```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1000  
t = np.linspace(0, 1, fs)
f1, f2 = 20, 60
sig = np.sin(2 * np.pi * f1 * t)  
sig += np.sin(2 * np.pi * f2 * t)  

scales = np.arange(1, 128)  
coef, freqs = pywt.cwt(sig, scales, 'morl')  

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

ax1.plot(t, sig)
ax1.set_ylabel('Amplitude')
cwt_plot = ax2.imshow(np.abs(coef), extent=[0, 1, 1, 128], cmap='jet', aspect='auto')
ax2.set_xlabel('Time')
ax2.set_ylabel('Scale')

plt.savefig('wavelet.png', bbox_inches='tight', transparent=True, dpi=500)

plt.tight_layout()
plt.show()
```
