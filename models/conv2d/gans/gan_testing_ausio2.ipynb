```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

from gan import GAN
```

Loading model.

```python
gan2 = GAN(device='cpu', latent_dim=2)
gan4 = GAN(device='cpu', latent_dim=4)
gan8 = GAN(device='cpu', latent_dim=8)
gan16 = GAN(device='cpu', latent_dim=16)
gan32 = GAN(device='cpu', latent_dim=32)
gan64 = GAN(device='cpu', latent_dim=64)
```

```python
model2 = torch.load('best_models/gan_ld2_ausio2_epoch200.pth', map_location=torch.device('cpu'), weights_only=True)
model4 = torch.load('best_models/gan_ld4_ausio2_epoch200.pth', map_location=torch.device('cpu'), weights_only=True)
model8 = torch.load('best_models/gan_ld8_ausio2_epoch200.pth', map_location=torch.device('cpu'), weights_only=True)
model16 = torch.load('best_models/gan_ld16_ausio2_epoch200.pth', map_location=torch.device('cpu'), weights_only=True)
model32 = torch.load('best_models/gan_ld32_ausio2_epoch200.pth', map_location=torch.device('cpu'), weights_only=True)
model64 = torch.load('best_models/gan_ld64_ausio2_epoch200.pth', map_location=torch.device('cpu'), weights_only=True)

gan2.generator.load_state_dict(model2['generator'])
gan2.discriminator.load_state_dict(model2['discriminator'])
gan4.generator.load_state_dict(model4['generator'])
gan4.discriminator.load_state_dict(model4['discriminator'])
gan8.generator.load_state_dict(model8['generator'])
gan8.discriminator.load_state_dict(model8['discriminator'])
gan16.generator.load_state_dict(model16['generator'])
gan16.discriminator.load_state_dict(model16['discriminator'])
gan32.generator.load_state_dict(model32['generator'])
gan32.discriminator.load_state_dict(model32['discriminator'])
gan64.generator.load_state_dict(model64['generator'])
gan64.discriminator.load_state_dict(model64['discriminator'])
```

## Learning Curves

Plotting learning curves.

```python
loss_g2 = np.load('best_models/loss_gan_ld2_g_ausio2.npy')
loss_d2 = np.load('best_models/loss_gan_ld2_d_ausio2.npy')
loss_g4 = np.load('best_models/loss_gan_ld4_g_ausio2.npy')
loss_d4 = np.load('best_models/loss_gan_ld4_d_ausio2.npy')
loss_g8 = np.load('best_models/loss_gan_ld8_g_ausio2.npy')
loss_d8 = np.load('best_models/loss_gan_ld8_d_ausio2.npy')
loss_g16 = np.load('best_models/loss_gan_ld16_g_ausio2.npy')
loss_d16 = np.load('best_models/loss_gan_ld16_d_ausio2.npy')
loss_g32 = np.load('best_models/loss_gan_ld32_g_ausio2.npy')
loss_d32 = np.load('best_models/loss_gan_ld32_d_ausio2.npy')
loss_g64 = np.load('best_models/loss_gan_ld64_g_ausio2.npy')
loss_d64 = np.load('best_models/loss_gan_ld64_d_ausio2.npy')

epochs2 = [i for i in range(loss_g2.shape[0])]
epochs4 = [i for i in range(loss_g4.shape[0])]
epochs8 = [i for i in range(loss_g8.shape[0])]
epochs16 = [i for i in range(loss_g16.shape[0])]
epochs32 = [i for i in range(loss_g32.shape[0])]
epochs64 = [i for i in range(loss_g64.shape[0])]
```

```python
plt.plot(epochs2, loss_g2, label='G2', c='blue')
plt.plot(epochs2, loss_d2, label='D2', c='blue', linestyle='--')
plt.plot(epochs4, loss_g4, label='G4', c='red')
plt.plot(epochs4, loss_d4, label='D4', c='red', linestyle='--')
plt.plot(epochs8, loss_g8, label='G8', c='orange')
plt.plot(epochs8, loss_d8, label='D8', c='orange', linestyle='--')
plt.plot(epochs16, loss_g16, label='G16', c='green')
plt.plot(epochs16, loss_d16, label='D16', c='green', linestyle='--')
plt.plot(epochs32, loss_g32, label='G32', c='purple')
plt.plot(epochs32, loss_d32, label='D32', c='purple', linestyle='--')
plt.plot(epochs64, loss_g64, label='G64', c='yellow')
plt.plot(epochs64, loss_d64, label='D64', c='yellow', linestyle='--')

plt.ylabel('Cross Entropy')
plt.xlabel('Epochs')
plt.title('Learning Curves')
plt.legend()
plt.show()
```

## Model avaliation

Loading data.

```python
data = np.load('tipAu_SiO2_morl.npz')['wavelets']

data = torch.tensor(data, dtype=torch.float32).unsqueeze(1)
data /= data.max()
```

Testing.

```python
sample2 = torch.randn(1, 2)
sample4 = torch.randn(1, 4)
sample8 = torch.randn(1, 8)
sample16 = torch.randn(1, 16)
sample32 = torch.randn(1, 32)
sample64 = torch.randn(1, 64)

with torch.no_grad():
    gen2 = gan2.generator(sample2)
    gen4 = gan4.generator(sample4)
    gen8 = gan8.generator(sample8)
    gen16 = gan16.generator(sample16)
    gen32 = gan32.generator(sample32)
    gen64 = gan64.generator(sample64)
    
fig, ax = plt.subplots(7, 1, figsize=(20, 12))

im0 = ax[0].imshow(gen2[0].squeeze())
ax[0].axis('off')
ax[0].set_title(f'LD=2')
divider = make_axes_locatable(ax[0])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0, cax=cax)

im0 = ax[1].imshow(gen4[0].squeeze())
ax[1].axis('off')
ax[1].set_title(f'LD=4')
divider = make_axes_locatable(ax[1])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0, cax=cax)

im0 = ax[2].imshow(gen8[0].squeeze())
ax[2].axis('off')
ax[2].set_title(f'LD=8')
divider = make_axes_locatable(ax[2])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0, cax=cax)

im0 = ax[3].imshow(gen16[0].squeeze())
ax[3].axis('off')
ax[3].set_title(f'LD=16')
divider = make_axes_locatable(ax[3])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0, cax=cax)

im0 = ax[4].imshow(gen32[0].squeeze())
ax[4].axis('off')
ax[4].set_title(f'LD=32')
divider = make_axes_locatable(ax[4])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0, cax=cax)

im0 = ax[5].imshow(gen64[0].squeeze())
ax[5].axis('off')
ax[5].set_title(f'LD=64')
divider = make_axes_locatable(ax[5])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0, cax=cax)

im0 = ax[6].imshow(data[0].squeeze())
ax[6].axis('off')
ax[6].set_title(f'Real')
divider = make_axes_locatable(ax[6])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0, cax=cax)

plt.tight_layout()
plt.show()
```
