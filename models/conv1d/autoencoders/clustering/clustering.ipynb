```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import adjusted_rand_score, homogeneity_score, completeness_score
from sklearn.decomposition import PCA
```

## LD=2

Loading data.

```python
ld2_ligsio2 = np.load('../best_models/ld2_ligsio2_all.npy')
ld2_ligcl = np.load('../best_models/ld2_ligcl_all.npy')
ld2_ausio2 = np.load('../best_models/ld2_ausio2_all.npy')
ld2_auau = np.load('../best_models/ld2_auau_all.npy')
```

Stacking the data together and labeling the datasets.

```python
ld2 = np.concatenate([ld2_ligsio2, ld2_ligcl, ld2_ausio2, ld2_auau], axis=0)

ld2_labels = (
    ['ligsio2'] * len(ld2_ligsio2) +
    ['ligcl']   * len(ld2_ligcl) +
    ['ausio2']  * len(ld2_ausio2) +
    ['auau']    * len(ld2_auau)
)
```

```python
colors = ['orange', 'green', 'blue', 'purple']
labels = ['LigSiO2', 'LigCl', 'AuSiO2', 'AuAu']

plt.figure(figsize=(8, 6))
start = 0
for i, (data, label, color) in enumerate(zip(
    [ld2_ligsio2, ld2_ligcl, ld2_ausio2, ld2_auau],
    labels,
    colors
)):
    plt.scatter(data[:, 0], data[:, 1], label=label, s=40, color=color)
plt.legend()
plt.title("LD=2")
plt.xlabel("LD 1")
plt.ylabel("LD 2")
plt.tight_layout()
plt.show()
```

Clustering LD=2 with Ward hierarchical clustering.

```python
linked_ld2 = linkage(ld2, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(
    linked_ld2, 
    truncate_mode='lastp',  
    p=20,                  
    leaf_rotation=45.,
    leaf_font_size=12.,
    show_contracted=True
)
plt.title('Dendrogram (Ward Linkage) LD=2')
plt.xlabel('Sample Index')
plt.ylabel('Distance (Ward)')
plt.show()
```

```python
cluster_labels_ld2 = fcluster(linked_ld2, t=4, criterion='maxclust')

plt.figure(figsize=(8, 6))
for cluster_id in np.unique(cluster_labels_ld2):
    mask = (cluster_labels_ld2 == cluster_id)
    plt.scatter(ld2[mask, 0], ld2[mask, 1], label=f"Cluster {cluster_id}", s=40, c=colors[cluster_id - 1])
plt.legend()
plt.title("Ward clustering LD=2")
plt.xlabel("LD 1")
plt.ylabel("LD 2")
plt.tight_layout()
plt.show()
```

```python
label_encoder = LabelEncoder()
y_true_encoded = label_encoder.fit_transform(ld2_labels)

cluster_encoder = LabelEncoder()
cluster_labels_num = cluster_encoder.fit_transform(cluster_labels_ld2)

ari = adjusted_rand_score(y_true_encoded, cluster_labels_ld2)
hom = homogeneity_score(y_true_encoded, cluster_labels_ld2)
comp = completeness_score(y_true_encoded, cluster_labels_ld2)

print(f"Ajusted Rand Index (ARI): {ari:.3f}")
print(f"Homogeneinity: {hom:.3f}")
print(f"Completeness: {comp:.3f}")
```

## LD=3

```python
ld3_ligsio2 = np.load('../best_models/ld3_ligsio2_all.npy')
ld3_ligcl = np.load('../best_models/ld3_ligcl_all.npy')
ld3_ausio2 = np.load('../best_models/ld3_ausio2_all.npy')
ld3_auau = np.load('../best_models/ld3_auau_all.npy')
```

```python
ld3 = np.concatenate([ld3_ligsio2, ld3_ligcl, ld3_ausio2, ld3_auau], axis=0)

labels_true_3 = (
    ['LigSiO2'] * len(ld3_ligsio2) +
    ['LigCL']   * len(ld3_ligcl) +
    ['AuSiO2']  * len(ld3_ausio2) +
    ['AuAu']    * len(ld3_auau)
)

label_encoder_3 = LabelEncoder()
y_true_3 = label_encoder_3.fit_transform(labels_true_3)
label_names_3 = label_encoder_3.classes_

linked_ld3 = linkage(ld3, method='ward')
cluster_labels_3 = fcluster(linked_ld3, t=4, criterion='maxclust')

pca = PCA(n_components=2)
ld3_pca = pca.fit_transform(ld3)
```

```python
colors = ['orange', 'green', 'blue', 'purple']
label_to_color = dict(zip(label_names_3, colors))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6)) 

for label in label_names_3:
    mask = (np.array(labels_true_3) == label)
    ax1.scatter(ld3_pca[mask, 0], ld3_pca[mask, 1], 
                label=label, s=40, color=label_to_color[label])
ax1.set_title("True Labels (PCA LD=3)")
ax1.set_xlabel("PC 1")
ax1.set_ylabel("PC 2")
ax1.legend()

for cluster_id in np.unique(cluster_labels_3):
    mask = (cluster_labels_3 == cluster_id)
    ax2.scatter(ld3_pca[mask, 0], ld3_pca[mask, 1], 
                label=f"Cluster {cluster_id}", s=40, c=colors[cluster_id - 1])
ax2.set_title("Ward Clustering (LD=3)")
ax2.set_xlabel("PC 1")
ax2.set_ylabel("PC 2")
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
dendrogram(
    linked_ld3, 
    truncate_mode='lastp',  
    p=20,                  
    leaf_rotation=45.,
    leaf_font_size=12.,
    show_contracted=True
)
plt.title('Dendrogram (Ward Linkage) LD=3')
plt.xlabel('Sample Index')
plt.ylabel('Distance (Ward)')
plt.tight_layout()
plt.show()
```

```python
cluster_encoder = LabelEncoder()
cluster_labels_num = cluster_encoder.fit_transform(cluster_labels_3)

ari_3 = adjusted_rand_score(y_true_3, cluster_labels_3)
homogeneity_3 = homogeneity_score(y_true_3, cluster_labels_3)
completeness_3 = completeness_score(y_true_3, cluster_labels_3)

print(f"Adjusted Rand Index (ARI):, {ari_3:.3f}")
print(f"Homogeneity: {homogeneity_3:.3f}")
print(f"Completeness: {completeness_3:.3f}")
```
